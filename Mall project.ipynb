```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Mall_Customers.csv")
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Age"], bins=8, color="skyblue", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Gender", data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Annual Income (k$)"], bins=8, color="orange", edgecolor="black")
plt.title("Annual Income Distribution")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Spending Score (1-100)"], bins=8, color="green", edgecolor="black")
plt.title("Spending Score Distribution")
plt.xlabel("Spending Score")
plt.ylabel("Number of Customers")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Annual Income (k$)"], df["Spending Score (1-100)"], color="red")
plt.title("Annual Income vs Spending Score")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Age"])
plt.title("Age Boxplot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Annual Income (k$)"])
plt.title("Annual Income Boxplot")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(y=df["Spending Score (1-100)"])
plt.title("Spending Score Boxplot")
plt.show()
```

```python
df_heat = df.copy()

df_heat["Gender"] = df_heat["Gender"].map({"Male":0,"Female":1})

plt.figure(figsize=(6,4))
sns.heatmap(df_heat.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
df.to_csv("Mall_Customers_Cleaned.csv", index=False)

print("Dataset Saved Successfully")
```
